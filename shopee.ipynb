```
data_email=data[['Id','Email']]
data_phone=data[['Id','Phone']]
data_order=data[['Id','OrderId']]
```

```
def drop_na(d):
  return d.dropna()
data_email=drop_na(data_email)
data_phone=drop_na(data_phone)
data_order=drop_na(data_order)
```

```
l1=data[data['Email']!=''].merge(data_email, on='Email', how='left' )
l1=l1[l1['Id_x']!=l1['Id_y']][['Id_x','Id_y']]
```

```
l2=data[data['Phone']!=''].merge(data_phone, on='Phone', how='left' )
l2=l2[l2['Id_x']!=l2['Id_y']][['Id_x','Id_y']]
```

```

l3=data[data['OrderId']!=''].merge(data_order, on='OrderId', how='left' )
l3=l3[l3['Id_x']!=l3['Id_y']][['Id_x','Id_y']]
```

```
l4=l1.merge(l2, right_on='Id_x',left_on='Id_x', how='inner')
l5=l2.merge(l3, right_on='Id_x',left_on='Id_x', how='inner')
l6=l1.merge(l3, right_on='Id_x',left_on='Id_x', how='inner')
```

```
l1.merge(l2, on='Id_x', how='outer')
```

```
def check_distinct(x):
  S=set([])
  for name in x.columns:
    S=S.union(set([item[0] for item in x[[name]].values]))
  return S
check_distinct(l4)
```

```
def get_full_key_dict(x):
    S=set([])
    for name in x.columns:
      S=S.union(set([item[0] for item in x[[name]].values]))
    return S
def dict_123(x):
    dict_key=list(get_full_key_dict(x))
    ori_1=x.values
    d={key : [] for key in dict_key}
    for item in ori_1:
      if item[0] in d:
        for i in range(1,len(item)):
            d[item[0]].append(item[i])
      else: 
        d[item[0]]=[]
    return d
def dict_456(x):
    dict_key=list(get_full_key_dict(x))
    ori_1=x.values
    d={key : [] for key in dict_key}
    for item in ori_1:
      for i in range(len(item)):
          if item[i] in d:
            for j in range(len(item)):
                  d[item[i]].append(item[j])
          else: 
            d[item[i]]=[]
    return d
def dict_transform(x):
    d={key : [] for key, value in x.items()}
    for key, value in x.items():
        d[key]=value
        d[key].append(key)
        d[key]=list(set(d[key]))
    return d

check_distinct(l4)

    
dl1=dict_transform(dict_123(l1))
dl2=dict_transform(dict_123(l2))
dl3=dict_transform(dict_123(l3))
dl4=dict_transform(dict_456(l4))
dl5=dict_transform(dict_456(l5))
dl6=dict_transform(dict_456(l6))
```
